```python
import json
import pandas as pd
# https://api.snowobs.com/v1/station/current?token=6628942ac4c3ca89c9989584aa160cd9&units=english&source=mesowest,snotel
# from: https://www.sierraavalanchecenter.org/weather-station-map#weather-map-table
```

```python
with open('./data/weather_stations.json', 'r') as f:
    data = json.load(f)
```

```python
data_matrix = []
for station in data['station_current']['STATION']:
    data_matrix.append([station['NAME'], float(station['ELEVATION']), station['STATUS'], station['SOURCE']])
    # ,  station['OBSERVATIONS'].keys()
```

```python
df = pd.DataFrame(data_matrix, columns=['name', 'elevation', 'status', 'source'])
```
